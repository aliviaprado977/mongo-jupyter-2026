```python
consulta_valor = list(
    colecao_atendimentos.find(
        {"valor": {"$gte": 90, "$lte": 200}},
        {"_id": 0}
    )
)

print("Quantidade encontrada:", len(consulta_valor))
consulta_valor[:5]
```

```python
consulta_cidades = list(
    colecao_atendimentos.find(
        {"cidade": {"$in": ["Assis", "Ourinhos", "Pompeia", "Marília"]}},
        {"_id": 0}
    )
)

print("Quantidade encontrada:", len(consulta_cidades))
consulta_cidades[:5]
```

```python
df = pd.DataFrame(consulta_cidades)
df.head()
```

```python
valores_nulos = df.isnull().sum()
print(valores_nulos)
```

```python
faturamento_total = float(df["valor"].sum())

faturamento_por_cidade = df.groupby("cidade")["valor"].sum().to_dict()
faturamento_por_cidade = {k: float(v) for k, v in faturamento_por_cidade.items()}

faturamento_por_servico = df.groupby("servico")["valor"].sum().to_dict()
faturamento_por_servico = {k: float(v) for k, v in faturamento_por_servico.items()}

ticket_medio_geral = float(df["valor"].mean())

ticket_medio_por_cidade = df.groupby("cidade")["valor"].mean().to_dict()
ticket_medio_por_cidade = {k: float(v) for k, v in ticket_medio_por_cidade.items()}

servico_mais_lucrativo = df.groupby("servico")["valor"].sum().idxmax()
cidade_maior_faturamento = df.groupby("cidade")["valor"].sum().idxmax()

top_3_atendimentos = df.nlargest(3, "valor").to_dict(orient="records")
```

```python
relatorio = {
    "data_geracao": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    "faturamento_total": faturamento_total,
    "faturamento_por_cidade": faturamento_por_cidade,
    "faturamento_por_servico": faturamento_por_servico,
    "ticket_medio_geral": ticket_medio_geral,
    "ticket_medio_por_cidade": ticket_medio_por_cidade,
    "servico_mais_lucrativo": servico_mais_lucrativo,
    "cidade_maior_faturamento": cidade_maior_faturamento,
    "top_3_atendimentos": top_3_atendimentos
}

relatorio
```

```python
colecao_relatorios.insert_one(relatorio)
print("Relatório salvo com sucesso na coleção relatorios.")
```
